```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img1 = Image.open('images/original/original1.jpg')
img2 = Image.open('images/original/original2.jpg')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
def kmeans_segmentation(img, clusters):

    img = np.asarray(img)/255.
    img_n = img.reshape(img.shape[0]*img.shape[1], img.shape[2])

    kmeans = KMeans(n_clusters=clusters, random_state=0).fit(img_n)
    imgshow = kmeans.cluster_centers_[kmeans.labels_]
    img = imgshow.reshape(img.shape[0], img.shape[1], img.shape[2])

    return img
```

```python
new_img1 = kmeans_segmentation(img1, 2)
plt.imshow(new_img1)
plt.imsave('images/segmentation/kmeans_segmentation1.jpg', new_img1)
```

```python
new_img2 = kmeans_segmentation(img2, 2)
plt.imshow(new_img2)
plt.imsave('images/segmentation/kmeans_segmentation2.jpg', new_img2)
```
