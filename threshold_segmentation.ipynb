```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = Image.open('images/original/original1.jpg')
img2 = Image.open('images/original/original2.jpg')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1 = img1.convert('L')
img2 = img2.convert('L')
```

```python
plt.imshow(img1,cmap='gray')
```

```python
plt.imshow(img2,cmap='gray')
```

```python
def threshold_segmentation(img):
    
    img = np.asarray(img)/255.
    threshold = np.mean(img)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            pixel = img[i][j]
            if(pixel < threshold):
                img[i][j] = 0
            else:
                img[i][j] = 1
    
    return img
```

```python
new_img1 = threshold_segmentation(img1)
plt.imshow(new_img1, cmap='gray')
plt.imsave('images/segmentation/threshold_segmentation1.jpg', new_img1, cmap='gray')
```

```python
new_img2 = threshold_segmentation(img2)
plt.imshow(new_img2, cmap='gray')
plt.imsave('images/segmentation/threshold_segmentation2.jpg', new_img2, cmap='gray')
```
